## Zadanie 1
### Podpunkt a
Korzystając z biblioteki NumPy, wygeneruj 500 liczb losowych z zakresu [0,100], używając w tym celu funkcji numpy.random.uniform.

```python
random_uniform = np.random.uniform(0,100, 500)
```

```python
print(random_uniform)
```

### Podpunkt b
Następnie, przy pomocy biblioteki Matplotlib, zaprezentuj utworzone dane na wy- kresie punktowym (matplotlib.pyplot.plot lub matplotlib.pyplot.scatter), gdzie oś X przedstawia indeksy danych, a oś Y – ich wartości; punkty danych przed- staw jako czerwone kropki.

```python
mpl.scatter(range(0, len(random_uniform)), random_uniform, color='red')
mpl.xlabel("Index")
mpl.ylabel("Value")
mpl.show()
```

### Podpunkt c
Dla tych samych danych wygeneruj histogram o 25 przedziałach (parametr bin=25). W tym celu zapoznaj się ze składnią funkcji matplotlib.pyplot.hist, a następ- nie napisz odpowiedni kod. Zadbaj o odpowiednie skale i oznaczenia na wykresie, legendę oraz ogólną czytelność.


```python
mpl.hist(random_uniform, bins=25)
mpl.xlabel("Value ranges")
mpl.ylabel("Frequency")
mpl.show()
```
